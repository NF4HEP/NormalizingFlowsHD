```python
def Extract(list,element):
    return [item[element] for item in list]

# open file and create dictionary with keys
a_file = open("./log_file_eternal_MsplineN_final_uncorrelated.txt")
my_dict = dict.fromkeys(a_file.readline().split(','), [])

# create a list of lists with all the information in the file
my_list = []
for line in a_file:
    splitted_line = line.split(',')
    array = []
    array.extend(splitted_line)
    my_list.append(array)

# fill the dictionary using the keys already defined and the information in the list of lists 
i=0
for key in my_dict:
    #print(key+' '+str(i))
    my_dict[key] = [Extract(my_list,i)]
    i=i+1

# print the filled dictionary
print(my_dict)
```

```python
df=pd.DataFrame.from_dict(my_dict).transpose()
pd.set_option('display.max_colwidth', 50)
left_df = df.style.set_properties(**{'text-align': 'left'})
display(left_df)
```

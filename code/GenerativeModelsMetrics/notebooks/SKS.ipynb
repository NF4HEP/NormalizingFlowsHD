## Useful functions

```python
def MixtureGaussian(ncomp: int,
                    ndims: int,
                    seed: int = 0) -> tfp.distributions.Mixture:
    targ_dist: tfp.distributions.Mixture = MixtureDistributions.MixMultiNormal1(ncomp,ndims,seed=seed)
    return targ_dist
```

```python
def central_moment(x, n):
    """Calculate the n-th central moment of a sample x."""
    mean_x = np.mean(x)
    return np.mean((x - mean_x)**n)
```

```python
@tf.function(jit_compile=True, reduce_retracing = True)
def sks_2samp_tf(data1: tf.Tensor, 
                 data2: tf.Tensor,
                 nslices: int = 100,
                 cov: bool = False,
                ) -> Tuple[tf.Tensor, tf.Tensor, tf.Tensor]:
    """
    Compute the sliced Kolmogorov-Smirnov statistic between two samples.
    """
    @tf.function(jit_compile=True, reduce_retracing = True)
    def simplified_ks_2samp_tf(a, b):
        return GMetrics.ks_metrics.ks_2samp_tf(a, b, alternative='two-sided', method='auto', precision=100)[0]
        #return ks_2samp(a, b, alternative='two-sided', mode='asymp')[0]
    
    # Your existing code
    # Compute ndims
    ndims = tf.shape(data1)[1]
    
    # Generate random directions
    directions = tf.random.normal((nslices, ndims))
    directions /= tf.norm(directions, axis=1)[:, None]
    directions = tf.cast(directions, dtype=data1.dtype)
    
    # Compute projections for all directions at once
    data1_proj = tf.tensordot(data1, directions, axes=[[1],[1]])
    data2_proj = tf.tensordot(data2, directions, axes=[[1],[1]])
        
    if cov:
        data1_cov = tfp.stats.covariance(data1_proj, sample_axis=0, event_axis=-1)
        data2_cov = tfp.stats.covariance(data2_proj, sample_axis=0, event_axis=-1)
    else:
        data1_cov = None
        data2_cov = None
        
    # Transpose the projection tensor to have slices on the first axis
    data1_proj = tf.transpose(data1_proj)
    data2_proj = tf.transpose(data2_proj)
    
    # Apply wasserstein_distance to each slice using tf.vectorized_map
    sks_proj = tf.vectorized_map(lambda args: simplified_ks_2samp_tf(*args), (data1_proj, data2_proj))
    
    # Computing correlation matrix of the projections
    # corr_mat = tfp.stats.correlation(directions)
    
    # Compute mean and std
    sks_mean = tf.reduce_mean(sks_proj)
    sks_std = tf.math.reduce_std(sks_proj)

    return sks_mean, sks_std, sks_proj, data1_cov, data2_cov # type: ignore
```
